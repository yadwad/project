```python
from states import states
import pandas as pd
import json
# set low_memory=False
```

```python
df_winner = pd.read_csv("Map/city_all.csv")
states_list = list(df_winner['state_abbr'])
df_winner = df_winner[df_winner.state_abbr != 'IV'].reset_index()
df_winner = df_winner.drop('index', 1)
df_winner

cities_new= df_winner[["state_abbr","city_name","metric_name","est"]]
cities_new

cities_new.rename(columns = {'metric_name':'metric_name_state','city_name':'city_name_state'}, inplace = True)
cities_new
```

```python
final_geojson = {"type": "FeatureCollection","features":[]}
for state in states_list:
    data = json.load(open("us_territories_geojson/{}/shape.geojson".format(state)))
    dict_add = {"type":"Feature",
                "properties":{"state_name":"{}".format(states[state]), 
                              "city":"{}".format(cities_new[cities_new['state_abbr']==state]['city_name_state'].item()), 
                              "metric":"{}".format(cities_new[cities_new['state_abbr']==state]['metric_name_state'].item()),                    
                              "estimate":"{}".format(cities_new[cities_new['state_abbr']==state]['est'].item())},
                "geometry": data
               }
    print(states[state] + " done")
    final_geojson['features'].append(dict_add)
```
